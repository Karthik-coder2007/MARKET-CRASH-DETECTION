```python
# Calculate daily returns
df['Returns'] = df['Close'].pct_change()

# Drop missing values
df.dropna(inplace=True)

df.head()
```

```python
def cusum_detection(returns, threshold=0.02, drift=0.001):
    """
    CUSUM for detecting downward market crashes
    """
    s_neg = [0]
    signals = [False]

    mean = returns.mean()

    for i in range(1, len(returns)):
        r = returns.iloc[i]

        s_new = min(0, s_neg[-1] + (r - mean - drift))
        s_neg.append(s_new)

        # Crash signal
        if s_new < -threshold:
            signals.append(True)
        else:
            signals.append(False)

    return np.array(s_neg), np.array(signals)
```

```python
df['CUSUM'], df['Crash'] = cusum_detection(df['Returns'])

print("Total crashes detected:", df['Crash'].sum())
```

```python
plt.figure(figsize=(14,6))

# Plot price
plt.plot(df['Date'], df['Close'], label='S&P 500 Price')

# Highlight crashes
crashes = df[df['Crash'] == True]
plt.scatter(crashes['Date'], crashes['Close'],
            color='red', label='Crash Detected')

plt.title("Market Crash Detection using CUSUM")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14,5))

plt.plot(df['Date'], df['CUSUM'], label='CUSUM')
plt.axhline(-0.02, color='red', linestyle='--', label='Threshold')

plt.title("CUSUM Chart (Crash Detection)")
plt.legend()
plt.show()
```

```python
crash_dates = df[df['Crash'] == True][['Date', 'Close']]
crash_dates.tail(10)
```

```python
# Filter only strong crash signals (avoid too many points)
major_crashes = df[(df['Crash'] == True) & (df['Returns'] < -0.02)]

plt.figure(figsize=(14,6))
plt.plot(df['Date'], df['Close'], label='Price')

plt.scatter(major_crashes['Date'], major_crashes['Close'],
            color='red', label='Major Crash')

plt.title("Major Crash Detection (Filtered)")
plt.legend()
plt.show()
```

```python
df['Month'] = df['Date'].dt.to_period('M')

monthly_crashes = df[df['Crash'] == True].groupby('Month').size()

monthly_crashes.plot(kind='bar', figsize=(12,5))
plt.title("Monthly Crash Frequency")
plt.show()
```

```python
df['Volatility'] = df['Returns'].rolling(20).std()

plt.figure(figsize=(12,5))
plt.plot(df['Date'], df['Volatility'])
plt.title("Market Volatility")
plt.show()
```

```python
df['Target'] = df['Crash'].astype(int)
```

```python
vol_threshold = df['Volatility'].mean() + df['Volatility'].std()

df['Risk_Level'] = "Low"

df.loc[df['Crash'], 'Risk_Level'] = "Medium"
df.loc[(df['Crash']) & (df['Volatility'] > vol_threshold), 'Risk_Level'] = "High"

df[['Date','Risk_Level']].tail()
```

```python
# Download NASDAQ data
nasdaq = yf.download("^IXIC", start="2010-01-01", end="2024-01-01")

nasdaq.reset_index(inplace=True)

# Calculate returns
nasdaq['Returns'] = nasdaq['Close'].pct_change()

nasdaq.dropna(inplace=True)
```

```python
# Apply CUSUM on NASDAQ
nasdaq['CUSUM'], nasdaq['Crash'] = cusum_detection(nasdaq['Returns'])

print("NASDAQ crashes:", nasdaq['Crash'].sum())
```

```python
plt.figure(figsize=(14,6))

# Plot both indices
plt.plot(df['Date'], df['Close'], label='S&P 500')
plt.plot(nasdaq['Date'], nasdaq['Close'], label='NASDAQ')

plt.title("Market Comparison")
plt.legend()
plt.show()
```
